```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/parkinsons.data')
df
```

```python
df = df.drop(['name'], axis = 1)
df.head()
```

```python
df['status'] = df['status'] ^ 1
```

```python
sns.set(style="darkgrid")
sns.barplot(x="Class", y="cnt", data = {"Class": ['negative', 'positive'], "cnt": [len(df[df['status'] == 0]), len(df[df['status'] == 1])]})
plt.title("Class distribution")
plt.xlabel("Classes")
plt.ylabel("Frequency")
plt.show()
```

```python
df.to_csv('../data/parkinsons.csv')
```
